```python
# remove unnecessary columns to analyse the df
#remove columns DATE_DIED, PNEUMONIA , ICU
columns_to_remove = ['DATE_DIED', 'PNEUMONIA', 'ICU', 'INTUBED', 'COPD' ,'INMSUPR', 'OTHER_DISEASE' ,'CARDIOVASCULAR','TOBACCO' , 'OBESITY', 'ASTHMA','DIABETES' ]
# Create a copy of the original DataFrame
df_covid_data_clean = df_covid_data.copy()
# Remove the specified columns in the copied DataFrame
df_covid_data_clean.drop(columns=columns_to_remove, inplace=True)
df_covid_data_clean
```

# Does Hypertension have an impact on Covid cases and if yes does it matter with age?

```python

# Filter the DataFrame for HIPERTENSION values of 1
hipertension_yes = df_covid_data_clean[df_covid_data_clean['HIPERTENSION'] == 1]
filtered_df_yes = hipertension_yes[hipertension_yes['CLASIFFICATION_FINAL'].between(1, 3)]

# Filter the DataFrame for HIPERTENSION values of 2
hipertension_no = df_covid_data_clean[df_covid_data_clean['HIPERTENSION'] == 2]
filtered_df_no = hipertension_no[hipertension_no['CLASIFFICATION_FINAL'].between(1, 3)]

# Count the occurrences of each 'CLASIFFICATION_FINAL' value for HIPERTENSION=1
classification_counts_yes = filtered_df_yes['CLASIFFICATION_FINAL'].value_counts()

# Count the occurrences of each 'CLASIFFICATION_FINAL' value for HIPERTENSION=2
classification_counts_no = filtered_df_no['CLASIFFICATION_FINAL'].value_counts()

# Create x-axis positions for the bars
x_positions_yes = np.arange(len(classification_counts_yes))
x_positions_no = np.arange(len(classification_counts_no)) + 0.4  # Add an offset to separate the bars

# Create a bar chart with both sets of data
plt.figure(figsize=(10, 6))
plt.bar(x_positions_yes, classification_counts_yes, width=0.4, label='Hipertension Yes', color='red')
plt.bar(x_positions_no, classification_counts_no, width=0.4, label='Hipertension No', color='orange')

# Set x-axis labels and title
plt.xlabel('CLASIFFICATION_FINAL')
plt.ylabel('Count')
plt.title('Comparison of CLASIFFICATION_FINAL for Hipertension (1 and 2, CLASIFFICATION_FINAL 1-3)')

# Set x-axis ticks and labels
plt.xticks(x_positions_yes + 0.2, classification_counts_yes.index)

# Display legend
plt.legend()

plt.show()
```

```python

# Combine the data for 'HIPERTENSION' values of 1 and 2
combined_df = pd.concat([filtered_df_yes, filtered_df_no])

# Create a box plot for 'AGE' with both 'HIPERTENSION' values
plt.figure(figsize=(12, 6))

# Box plot for 'AGE' with 'CLASIFFICATION_FINAL' and both 'HIPERTENSION' values
sns.boxplot(x='CLASIFFICATION_FINAL', y='AGE', data=combined_df, hue='HIPERTENSION', width=0.4, palette='Set2', dodge=True)

# Set labels and title
plt.xlabel('CLASIFFICATION_FINAL')
plt.ylabel('Age')
plt.title('Boxplot of Age for CLASIFFICATION_FINAL (1-3) - Both Hipertension Values')

# Display legend
plt.legend(title='HIPERTENSION', loc='upper right')

plt.show()
```

```python
# Create a new column 'Age_Group' to represent age in chunks of 5
combined_df['Age_Group'] = pd.cut(combined_df['AGE'], bins=range(0, 110, 5), right=False)

# Create a bar chart for each age group with 'HIPERTENSION' values
plt.figure(figsize=(12, 6))

# Bar chart for 'Age_Group' with 'HIPERTENSION' values
sns.countplot(x='Age_Group', hue='HIPERTENSION', data=combined_df, palette='Set2')

# Set labels and title
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Bar Chart of Age Groups with HIPERTENSION Values')

# Display legend
plt.legend(title='HIPERTENSION', loc='upper right')
plt.xticks(rotation=40)

plt.show()
```

# Does the Medical unit have any relevance to the symptoms of the patients or age?

```python
# Group by the "Medical Unit" column and count occurrences
grouped_medical_unit = df_covid_data.groupby('MEDICAL_UNIT').size()

# Display the resulting grouped DataFrame
print(grouped_medical_unit)
```

```python
# Create a new DataFrame without rows containing the value 98
df_without_98 = df_covid_data[~df_covid_data.isin([98]).any(axis=1)].copy()

# If you want to reset the index
df_without_98.reset_index(drop=True, inplace=True)

# Create a DataFrame with the counts for each combination of 'MEDICAL_UNIT', 'RENAL_CHRONIC', 'HIPERTENSION'
grouped_data = df_without_98.groupby(['MEDICAL_UNIT', 'RENAL_CHRONIC', 'HIPERTENSION']).size().reset_index(name='Count')

# Create a pivot table
pivot_table = pd.pivot_table(grouped_data, values='Count', index='MEDICAL_UNIT', columns=['RENAL_CHRONIC', 'HIPERTENSION'], aggfunc='sum', fill_value=0)

# Display the pivot table
pivot_table
```

```python
# Create a horizontal stacked bar chart
pivot_table.plot(kind='barh', stacked=True, colormap='viridis', figsize=(10, 6))

# Set labels and title
plt.xlabel('Count')
plt.ylabel('Medical Unit')
plt.title('Horizontal Stacked Bar Chart of Counts by Medical Unit')

# Show the plot
plt.show()
```
